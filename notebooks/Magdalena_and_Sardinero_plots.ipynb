```python
import pandas as pd
import hvplot.pandas  # This enables hvplot for pandas DataFrames
import panel as pn

# Load the dataset
historical_data = "/Users/ignasivalles/Google Drive/La meva unitat/POM/data/muestreoCubo1970_78.xlsx"
df = pd.read_excel(historical_data, header=0)

# Rename columns for consistency
df = df.rename(columns={
    'ao': 'year',
    'mes': 'month',
    'dia': 'day',
    'temperatura agua': 'Temperature'  #  Rename for clarity
})

# Create a new 'Date' column
df['Date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Filter data for Sardinero and Magdalena stations
sardinero = df[df['Numero estacion'] == 'Sardinero']
magdalena = df[df['Numero estacion'] == 'Magdalena']

# Create hvplot for Sardinero
sardinero_plot = sardinero.hvplot.line(
    x='Date', y='Temperature',  #  Updated column name
    title='Sea Temperature - Sardinero',
    xlabel='Year', ylabel='Temperature',
    grid=True, line_width=2, color='blue', label='Sardinero'
).opts(
    width=400,                 #  Reduced plot width
    height=200,              #  Set minimum width
    tools=['hover'],
    hover_tooltips=[('Date', '@Date{%F}'), ('Temperature', '@Temperature')],  #  Corrected tooltip
    margin=(0, 0, 0, 0)            #  Remove white spaces around the plot
)

# Create hvplot for Magdalena
magdalena_plot = magdalena.hvplot.line(
    x='Date', y='Temperature',  #  Updated column name
    title='Sea Temperature - Magdalena',
    xlabel='Year', ylabel='Temperature',
    grid=True, line_width=2, color='orange', label='Magdalena'
).opts(
    width=400,                 #  Reduced plot width
    height=200,                #  Set minimum width
    tools=['hover'],
    hover_tooltips=[('Date', '@Date{%F}'), ('Temperature', '@Temperature')],  #  Corrected tooltip
    margin=(0, 0, 0, 0)            #  Remove white spaces around the plot
)

# Display the plots interactively
pn.extension()
layout = pn.Column(sardinero_plot, magdalena_plot, sizing_mode='stretch_width')  #  Apply sizing_mode here

#  Save each plot as an HTML file without white spaces
pn.panel(sardinero_plot).save("/Users/ignasivalles/Oceanography/IEO/projects/POM/pom.github.io/sardinero_plot.html", embed=True)
pn.panel(magdalena_plot).save("/Users/ignasivalles/Oceanography/IEO/projects/POM/pom.github.io/magdalena_plot.html", embed=True)

layout
```
