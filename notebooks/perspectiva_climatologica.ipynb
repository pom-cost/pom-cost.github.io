```python
import sys
sys.path.append('/Users/ignasivalles/Oceanography/IEO/projects/pom-cost.github.io/src')
from data_functions import *
import pandas as pd
import numpy as np
import glob
import os
from holoviews import opts
from bokeh.models import HoverTool
import matplotlib.pyplot as plt
from holoviews import render
from bokeh.layouts import layout
from bokeh.resources import CDN
from bokeh.embed import file_html
from holoviews import save
hv.extension('bokeh')
```

```python
def limit_zoom(plot, element):
    plot.state.x_range.bounds = (0, 12)
    plot.state.y_range.bounds = (10, 25)

historical_data = "/Users/ignasivalles/Google Drive/La meva unitat/POM/data/muestreoCubo1970_78.xlsx"
all_data = '/Users/ignasivalles/Oceanography/IEO/projects/POM/pom-cost.github.io/data/individual_data.csv'

clim_plot = plot_climatological_year(historical_data)
df = pd.read_csv(all_data)
df = df.where(df.Temperature > 10)
obs_plot = plot_obs(df)
rolling_plot = plot_rolling_by_year(df)

combined_plot = clim_plot * obs_plot * rolling_plot

month_boundaries = np.arange(0, 13, 1)
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan']

shared_opts = dict(
    title="Climatological year",
    xticks=[(i, label) for i, label in zip(month_boundaries, month_labels)],
    show_grid=True,
    xlim=(0, 12),
    ylim=(5, 28),
    hooks=[limit_zoom],
)

combined_plot
```

```python
# Mobile version: responsive width, compact height
mobile_plot = combined_plot.opts(**shared_opts, responsive=True, height=280, margin=(0, 0, 0, 0), clone=True)
hv.save(mobile_plot,
        '/Users/ignasivalles/Oceanography/IEO/projects/pom-cost.github.io/climatological-year/climatological_plot_mbl.html',
        backend='bokeh')
print("Mobile plot saved.")
```
